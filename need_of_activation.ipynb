```python
import pandas as pd
import numpy as np

import tensorflow
from keras.layers import Dense
from keras.models import Sequential
from keras.optimizers import Adam

from sklearn.datasets import make_moons
import matplotlib.pyplot as plt
from mlxtend.plotting import plot_decision_regions
```

```python
X , y = make_moons(200, noise = 0.35, random_state = 42)
```

```python
plt.scatter(X[:, 0], X[: , -1] , c = y)
```

```python
model = Sequential()
model.add(Dense(128, activation = 'linear', input_dim = 2))
model.add(Dense(128, activation = 'linear'))
model.add(Dense(1, activation = 'sigmoid'))
```

```python
model.summary()
```

```python
adam = Adam(learning_rate = 0.01)
model.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
history = model.fit(X, y, epochs = 500, verbose = False, validation_split= 0.2)
```

```python
plot_decision_regions(X, y, clf = model)
```

```python
model1 = Sequential()
model1.add(Dense(128, activation = 'relu', input_dim = 2))
model1.add(Dense(128, activation = 'relu'))
model1.add(Dense(1, activation = 'sigmoid'))


adam = Adam(learning_rate = 0.01)
model1.compile(optimizer = adam, loss = 'binary_crossentropy', metrics = ['accuracy'])
history1 = model1.fit(X, y, epochs = 100, verbose = False, validation_split= 0.2)
```

```python
plot_decision_regions(X , y, clf = model1)
```

```python
plt.plot(history1.history['loss'])
plt.plot(history1.history['val_loss'])
```

```python
plt.figure(figsize = (10,5))
plt.subplot(121)
plt.title('without activation function')
plot_decision_regions(X, y, clf = model)

plt.subplot(122)
plt.title('After activation function')
plot_decision_regions(X , y, clf = model1)

plt.tight_layout()
plt.show()
```
